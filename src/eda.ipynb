```python
import numpy as np

# 1. Clipping (Winsorizing) der extremen Ausreißer
# Wir nutzen das 0.1% und 99.9% Quantil, um die wilden Millionen-Werte zu kappen
lower_bound = df['price_usd'].quantile(0.001)
upper_bound = df['price_usd'].quantile(0.999)

print(f"Setze harte Grenzen für price_usd: min={lower_bound:.2f}, max={upper_bound:.2f}")

# Neue Spalte für den geclippten Preis anlegen (Original bleibt für Checks erhalten)
df['price_usd_clipped'] = df['price_usd'].clip(lower=lower_bound, upper=upper_bound)


# 2. Log-Transformation anwenden
# Log(1+x) staucht die Verteilung und macht sie für das Modell leichter verdaulich
df['prop_log_price_usd'] = np.log1p(df['price_usd_clipped'])


# 3. Gruppeninterne Preis-Features bauen (Der wichtigste Teil!)
# Wir berechnen den Median-Preis für jede Suchanfrage (srch_id)
# transform() weist den Medianwert direkt wieder jeder Zeile der ursprünglichen Gruppe zu
group_median_price = df.groupby('srch_id')['price_usd_clipped'].transform('median')
group_mean_price = df.groupby('srch_id')['price_usd_clipped'].transform('mean')

# Feature A: Verhältnis zum Gruppen-Median (Ist das Hotel doppelt so teuer oder halb so teuer?)
# 1e-5 verhindert Division durch 0, falls der Median mal 0 sein sollte
df['price_group_median_ratio'] = df['price_usd_clipped'] / (group_median_price + 1e-5)

# Feature B: Differenz zum Gruppen-Mittelwert (Wie viel $ teurer/billiger als der Durchschnitt?)
# df['price_group_mean_diff'] = df['price_usd_clipped'] - group_mean_price

# Feature C: Preis-Rang innerhalb der Suche (Das billigste Hotel bekommt Rang 1)
df['price_rank_in_group'] = df.groupby('srch_id')['price_usd_clipped'].rank(method='min')


# Kurzer Check, wie die neuen Features für eine Beispiel-Suche aussehen
sample_srch_id = df['srch_id'].iloc[0]
check_cols = ['srch_id', 'price_usd', 'price_usd_clipped', 
              'prop_log_price_usd', 'price_group_median_ratio', 'price_rank_in_group']

print("\nErgebnis für eine Beispiel-Suche:")
print(df[df['srch_id'] == sample_srch_id][check_cols].sort_values('price_usd').head(10))
```

```python
# handle missing values + amo prop_location_score1 and 2 to categorical values as those are best for FM model
# Wir definieren Bins (z.B. 0-2, 2-4, 4-6, 6+)
bins = [-np.inf, 2, 4, 6, np.inf]
labels = ['low', 'medium', 'high', 'very_high']

# Werte in Buckets einteilen
df['location_score1_bucket'] = pd.cut(df['prop_location_score1'], bins=bins, labels=labels)
df['location_score2_bucket'] = pd.cut(df['prop_location_score2'], bins=bins, labels=labels)

# Die NaN-Werte füllen wir als EIGENE Kategorie auf
df['location_score1_bucket'] = df['location_score1_bucket'].cat.add_categories('missing')
df['location_score1_bucket'] = df['location_score1_bucket'].fillna('missing')

df['location_score2_bucket'] = df['location_score2_bucket'].cat.add_categories('missing')
df['location_score2_bucket'] = df['location_score2_bucket'].fillna('missing')
```

```python
# build other RANKING features: star_rank, price_diff_rank

# ---------------------------------------------------------
# 1. Star Rank (Sterne-Rang innerhalb der Suche)
# ---------------------------------------------------------
# Wichtig: Hier setzen wir ascending=False, denn mehr Sterne sind besser!
# Ein 5-Sterne Hotel bekommt also Rang 1, ein 3-Sterne Hotel einen schlechteren Rang.
# method='min' bedeutet: Wenn zwei Hotels 5 Sterne haben, bekommen beide Rang 1. 
# Das nächste Hotel (z.B. 4 Sterne) bekommt dann Rang 3.
df['star_rank'] = df.groupby('srch_id')['prop_starrating'].rank(method='min', ascending=False)


# ---------------------------------------------------------
# 2. Price Diff Rank (Rang des "Deals" / historischen Rabatts)
# ---------------------------------------------------------
# Zuerst berechnen wir die Differenz zwischen dem aktuellen Preis (den wir vorhin 
# logarithmiert haben) und dem historischen Preis des Hotels (der in den Daten 
# bereits logarithmiert ist!).
# Ein negativer Wert bedeutet hier: Das Hotel ist gerade billiger als historisch üblich (ein Schnäppchen!).
df['price_vs_historical_diff'] = df['prop_log_price_usd'] - df['prop_log_historical_price']

# Jetzt ranken wir diese Differenz innerhalb der Suchgruppe.
# ascending=True, weil der kleinste (stärkste negative) Wert das beste Schnäppchen ist 
# und somit Rang 1 bekommen soll.
df['price_diff_rank'] = df.groupby('srch_id')['price_vs_historical_diff'].rank(method='min', ascending=True)


# ---------------------------------------------------------
# Kurzer Check der Ergebnisse
# ---------------------------------------------------------
check_cols_rank = [
    'srch_id', 'prop_starrating', 'star_rank', 
    'prop_log_price_usd', 'prop_log_historical_price', 
    'price_vs_historical_diff', 'price_diff_rank'
]

# Wir schauen uns wieder die Beispiel-Suche von vorhin an
sample_srch_id = df['srch_id'].dropna().iloc[0]
print("\nErgebnis für Star-Rank und Price-Diff-Rank in einer Suche:")
print(df[df['srch_id'] == sample_srch_id][check_cols_rank].sort_values('price_diff_rank').head(10))
```

## split data

```python
from sklearn.model_selection import GroupShuffleSplit

# Wir reservieren 20% der Suchanfragen (nicht der einzelnen Zeilen!) für die Validierung
gss = GroupShuffleSplit(n_splits=1, test_size=0.2, random_state=42)

# gss.split spuckt die Indizes für Train und Val aus, gruppiert nach srch_id
train_idx, val_idx = next(gss.split(df, groups=df['srch_id']))

df_train = df.iloc[train_idx].copy()
df_val = df.iloc[val_idx].copy()

print(f"Anzahl Zeilen Train: {len(df_train)} | Anzahl Suchen Train: {df_train['srch_id'].nunique()}")
print(f"Anzahl Zeilen Val:   {len(df_val)} | Anzahl Suchen Val:   {df_val['srch_id'].nunique()}")
```

### create bin ID features

```python
def encode_frequency_ids(df_train, df_val, column_name, min_freq=10):
    """
    Erstellt ein "Wörterbuch" für häufige IDs basierend auf dem Trainings-Set 
    und wendet es sicher auf Train und Val an.
    """
    # 1. Zähle die Häufigkeit jeder ID NUR im Trainingsset
    value_counts = df_train[column_name].value_counts()
    
    # 2. Filtere nur die IDs heraus, die oft genug vorkommen
    frequent_ids = value_counts[value_counts >= min_freq].index
    
    # 3. Erstelle das Wörterbuch (alte_id -> neue_id)
    # Wir starten bei 1, denn die 0 ist für "Unknown/Rare" reserviert!
    id_mapping = {old_id: new_id for new_id, old_id in enumerate(frequent_ids, start=1)}
    
    # 4. Wende das Mapping auf das Trainings-Set an
    # map() macht seltene/unbekannte IDs zu NaN, fillna(0) macht sie zur Kategorie 0
    new_col_name = f"{column_name}_encoded"
    df_train[new_col_name] = df_train[column_name].map(id_mapping).fillna(0).astype(int)
    
    # 5. Wende EXAKT DASSELBE Mapping auf das Validation-Set an
    df_val[new_col_name] = df_val[column_name].map(id_mapping).fillna(0).astype(int)
    
    # Kurzes Feedback ausgeben
    print(f"{column_name}: {len(id_mapping)} IDs ins Wörterbuch aufgenommen (Rest ist Kategorie 0).")
    
    return df_train, df_val

# ==========================================
# Anwendung auf deine Daten
# ==========================================

# prop_id (Hotels): Hier gibt es extrem viele, setze min_freq ruhig etwas höher (z.B. 10 oder 20)
df_train, df_val = encode_frequency_ids(df_train, df_val, 'prop_id', min_freq=10)

# srch_destination_id (Reiseziele): Orte kommen oft häufiger vor, hier reicht vielleicht schon min_freq=5
df_train, df_val = encode_frequency_ids(df_train, df_val, 'srch_destination_id', min_freq=5)

# srch_room_count (Zimmeranzahl): Hat meist eh nur Werte wie 1, 2, 3, 4, 5. 
# Das können wir auch durchjagen, um Ausreißer (z.B. 10 Zimmer) in die 0 zu werfen
df_train, df_val = encode_frequency_ids(df_train, df_val, 'srch_room_count', min_freq=10)

# Kurzer Check im Validation Set, wie viele "Unknowns" (Kategorie 0) es gab:
print("\nVerteilung der encodierten prop_ids im Validation-Set:")
print(f"Kategorie 0 (Unknown/Rare): {(df_val['prop_id_encoded'] == 0).sum()} Zeilen")
print(f"Bekannte Kategorien (>0):   {(df_val['prop_id_encoded'] > 0).sum()} Zeilen")
```

```python
# problem: highly unbalanced data; this is a huge ass problem for the pointwise FM; therefore: throw away a huge part of the negative examples to better the ratio

import pandas as pd
import numpy as np

# 1. Gemeinsame Target-Variable erstellen (falls noch nicht passiert)
# Alles, was angeklickt ODER gebucht wurde, ist ein positives Signal (1)
df_train['target'] = np.where((df_train['click_bool'] == 1) | (df_train['booking_bool'] == 1), 1, 0)
df_val['target']   = np.where((df_val['click_bool'] == 1) | (df_val['booking_bool'] == 1), 1, 0)

print("Verteilung im Trainingsset VOR dem Downsampling:")
print(df_train['target'].value_counts(normalize=True) * 100) # In Prozent


# 2. Daten in positive und negative Klassen aufteilen
df_train_pos = df_train[df_train['target'] == 1]
df_train_neg = df_train[df_train['target'] == 0]

# 3. Wir legen das Verhältnis fest (z.B. 5 Nullen auf 1 Eins)
NEGATIVE_RATIO = 5
num_negatives_to_keep = len(df_train_pos) * NEGATIVE_RATIO

# 4. Downsampling per Zufall (sample)
# random_state sorgt dafür, dass das Ergebnis reproduzierbar bleibt
df_train_neg_sampled = df_train_neg.sample(n=num_negatives_to_keep, random_state=42)

# 5. Zusammenfügen und gut durchmischen!
# frac=1 zieht 100% der Zeilen in zufälliger Reihenfolge
df_train_balanced = pd.concat([df_train_pos, df_train_neg_sampled]).sample(frac=1, random_state=42).reset_index(drop=True)

print("\nVerteilung im Trainingsset NACH dem Downsampling:")
print(df_train_balanced['target'].value_counts())
print("\nNeues Verhältnis (in %):")
print(df_train_balanced['target'].value_counts(normalize=True) * 100)
```

```python
# first test of POINTWISE FM

from pointwise_fm import fit_encoders, transform, train_fm

# 1. Encoder NUR auf dem (downgesampelten) Train-Set fitten
encoders, field_dims = fit_encoders(df_train_balanced, n_bins=32)

# 2. beide Sets mit denselben Encodern in Integer-Codes umwandeln
X_tr, _ = transform(df_train_balanced, encoders)
X_va, _ = transform(df_val, encoders)

# 3. FM trainieren -- df_val geht komplett rein (braucht relevance + srch_id fuer NDCG)
y_tr = df_train_balanced["target"].to_numpy("float32")
model, best_ndcg = train_fm(X_tr, y_tr, X_va, df_val, field_dims)
```
